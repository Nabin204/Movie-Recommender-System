```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head(1)
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
movies.shape
```

```python
# Useful Columns
#genres
#id
#keywords
#title
#overview
#cast
#crew
```

```python
#we need this data in format like:
#['Action','Adventure','Fantasy','Scify']
import json
def convert(x):
    lst = []
    x=json.loads(x)
    for i in x:
        lst.append(i['name'])
    return lst
    
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'][0]
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
#For columns cast only
movies['cast'][0]
```

```python
def convert3(x):
    counter=0
    lst=[]
    x = json.loads(x)
    for i in x:
        if counter!=3:
            lst.append(i['name'])
            counter += 1
        else:
            break
    return lst
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
#For crew column
movies['crew'][0]
```

```python
def convert4(x):
    counter=0
    lst=[]
    x = json.loads(x)
    for i in x:
        if (counter!=3 and i['job']=='Director'):
            lst.append(i['name'])
            counter += 1
        else:
            pass
    return lst
    
```

```python
movies['crew']=movies['crew'].apply(convert4)
movies.head()
```

```python
movies['overview'].dtype
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split())
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags'] = movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
```

```python
movies.head()
```

```python
new_df = movies.drop(columns=['genres','keywords','cast','crew','overview'])
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x : " ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x : x.lower())
new_df.head()
```

```python
#Now we do stemming
#['danced','dancing'] -> ['danc','danc']
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
print(ps.stem('dancing'))
print(ps.stem('danced'))
print(ps.stem('running'))
print(ps.stem('runs'))
```

```python
def stemm(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stemm(new_df['tags'][0])
```

```python
new_df['tags']=new_df['tags'].apply(stemm)
new_df.head()
```

```python
# Now main code begins from here
# We have to convert each tags into vectors
# For this we will use countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
np.set_printoptions(threshold=np.inf)
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[-1])[1:6]
```

```python
#Main code
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    #While sorting, the index position of movies in similarity  matrix changes
    #To make it not lose, we use enumerate
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
